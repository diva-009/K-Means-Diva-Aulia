```python
import matplotlib.pyplot as plt

x = ["senin", "selasa", "rabu", "kamis", "jumat"]
y = [1, 2, 3, 4, 5]

warna_batang = ["black", "green"]
plt.plot(x, y, color= "red")
plt.bar(x, y, color = warna_batang)

plt.title("visualisasi data diva")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
